```python
import numpy as np
import matplotlib.pyplot as plt
```

# Line

```python
product1 = [10,20,30,40,50]
product2 = [15,30,10,20,60]
month = [1,2,3,4,5]
```

```python
plt.plot(month,product1,color='c',marker='o',linestyle='-.')
plt.plot(month,product2,'m|:') #เรียงลำดับ

plt.xlabel('Month',size=15,color='b',backgroundcolor='cyan')
# or
#data = {'size':20,'color':'b','backgroundcolor':'cyan'}
#plt.xlabel('Month',data)
plt.ylabel('Sales',size=15,color='r',backgroundcolor='yellow')
#plt.savefig('exportMPL.png',transparent=True) 
#ไม่เอาพื้นหลังสีขาว แต่ถ้าใช้.jpgจะยังมีbgสีขาวอยู่ดี
#plt.savefig('exportMPL_w_bg.png',transparent=False) 
#เอา white bg
configure = {'size':20,'color':'w','backgroundcolor':'k'}
plt.title('Top sell 2020',configure,loc='right')
plt.legend()
plt.show()
```

```python
data = [15,10,60,30,70]
label = ['Python','JavaScript','Java','PHP','HTML']
```

```python
plt.pie(data,labels=label)
plt.show()
```

```python
plt.pie(data,labels=label,autopct='%.2f%%')
plt.show()
```

```python
color = ['red','yellow','magenta','cyan','green']
exp = [0.2,0.4,0,0.35,0]
plt.pie(data,labels=label,colors=color,autopct='%.2f%%',shadow=True,explode=exp,startangle=-10)
plt.show()
```

Scatter plot

```python
x = [1,2,3,4,5]
y = [1.2,3,2.3,4.5,4.3]
size = [100,200,300,100,800]
color = ['r','g','b','c','m']
plt.scatter(x,y,s=size,c=color)
plt.show()
```

histogram

```python
age = [10,15,20,10,15,10,20,15,15,10,15,20,25,25,25,30,30,30,35,35,40,40,10]
plt.hist(age)
plt.xlabel('age')
plt.ylabel('count')
plt.title('Total Age histogram')
plt.legend(['Population'])
```

continuous histogram

```python
mean = 10
std = 2.5
age = np.random.normal(mean,std,5000)
plt.hist(age,bins=30) #30ช่วง
plt.show()
```

Bar

```python
course = ['python','java','c#']
score = [80,75,50]
color = ['r','orange','b']

plt.bar(course,score,color=color,width=0.6,edgecolor='k')
plt.xlabel('Course',fontsize='x-large')
plt.ylabel('score')
plt.title('Midterm Score')
plt.show()
```

```python
plt.barh(course,score,color=color,edgecolor='k',linewidth=5)
```

stack bar

```python
room = ['A','B','C']
boys = [15,10,5]
girls = [20,7,15]
plt.bar(room,boys,label='Boys',color='blue')
plt.bar(room,girls,bottom=boys,label='Girls',color='orange')
# bottom= เอาboys ไปอยู่ล่างgirls stack รวมเข้าด้วยกัน
```

```python
plt.stackplot(room,boys,girls,labels=['Boys','Girls'],colors=['green','pink'])
```

subplot

```python
def demo1():
  fig = plt.figure()
  fig.add_subplot(121)
  plt.plot([1,2,3,4,5],[1,2,3,2,5])
  fig.add_subplot(122)
  plt.bar(range(5),[3,5,2,4,2])
  plt.show()
demo1()
#พลอตแนวนอน
```

```python
#plot แนวตั้ง
def demo2():
  fig = plt.figure()
  fig.add_subplot(211) #2row1col firstpic
  plt.plot([1,2,3,4,5],[1,2,3,2,5])
  fig.add_subplot(212) #2row1col secpic
  plt.bar(range(5),[3,5,2,4,2])
  plt.show()
demo2()
```

```python
def demo3():
  fig = plt.figure()
  fig.add_subplot(221)
  plt.plot([1,2,3,4,5],[1,2,3,2,5])
  fig.add_subplot(222)
  plt.bar(range(5),[3,5,2,4,2])
  fig.add_subplot(223)
  plt.scatter([1,2,3,4,5],[1,2,3,2,5])
  plt.show()
demo3()
```

```python
def demo4():
  fig = plt.figure()
  fig.add_subplot(221)
  plt.plot([1,2,3,4,5],[1,2,3,2,5])
  fig.add_subplot(222)
  plt.bar(range(5),[3,5,2,4,2])
  fig.add_subplot(212) #ไม่เข้าใจเท่าไหร่
  plt.scatter([1,2,3,4,5],[1,2,3,2,5])
  plt.show()
demo4()
```

```python
def demo5():
  fig,ax = plt.subplots(2,3,figsize=(12,18))
  for i in range(ax.shape[0]):
    for j in range(ax.shape[1]):
      print(i,j)
      ax[i,j].plot(np.random.randint(1,11,5),np.random.randint(1,11,5)) #1-11เอา5ตัว
      ax[i,j].set_title(f'axis[{i},{j}]')
```
