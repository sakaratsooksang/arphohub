```python
import folium
import pandas as pd
from folium.plugins import MarkerCluster
# Function
def isNaN(num):
    if float('-inf') < float(num) < float('inf'):
        return False 
    else:
        return True
    
dat = pd.read_csv('/Users/sakarat/Desktop/Internship/Folium/data_point.csv')
name = dat['name'].values
lon = dat['lon'].values
lat = dat['lat'].values
df = pd.DataFrame({
    'name':name,
    'lon':lon,
    'lat':lat
})

m = folium.Map(location=[13.0389, 101.4901], zoom_start=12, tiles="Stamen Terrain")
marker_cluster = MarkerCluster().add_to(m)
for i in range(len(df)):
    if (not isNaN(df.lat[i])) and (df.lat[i]<20.33 and df.lon[i]<105):
        folium.Marker(
            [df.lat[i], df.lon[i]], popup="<i>Location : {} Coordinate : [{},{}]</i>".format(df.name[i],df.lat[i],df.lon[i]), tooltip= "Click to see Location"
        ).add_to(marker_cluster)
#x for x in range(20) if x % 2 == 0
latitude = [l for l in df.lat if not isNaN(l)  and (l<20.33)]
longitude = [l for l in df.lon if not isNaN(l) and (l<105) ]
# additional fucntion

#m.add_child(folium.ClickForMarker(popup="Waypoint"))
m.add_child(folium.LatLngPopup())
m.save("ThaiWebMap.html")
```
