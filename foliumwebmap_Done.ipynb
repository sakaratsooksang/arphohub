```python
import folium
import pandas as pd

# Function
def isNaN(num):
    if float('-inf') < float(num) < float('inf'):
        return False 
    else:
        return True
    
dat = pd.read_csv('/Users/sakarat/Desktop/Internship/Folium/data_point.csv')
name = dat['name'].values
lon = dat['lon'].values
lat = dat['lat'].values
df = pd.DataFrame({
    'name':name,
    'lon':lon,
    'lat':lat
})

m = folium.Map(location=[13.0389, 101.4901], zoom_start=12, tiles="Stamen Terrain")
for i in range(len(df)):
    if (not isNaN(df.lat[i])) and (df.lat[i]<20.33 and df.lon[i]<105):
        folium.Marker(
            [df.lat[i], df.lon[i]], popup="<i> {} </i>".format(df.name[i]), tooltip= "Click to see Location"
        ).add_to(m)
m.save("ThaiWebMap.html")
```
